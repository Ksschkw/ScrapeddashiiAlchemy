```python
print("Do the 41 conditions in SymbiPredict overlap with the 116 in DSAPP? checking merging compatibility")
symb_diseases = set(symb['prognosis'].unique())
dsapp_diseases = set(dsapp['Disease'].unique())
common_diseases = symb_diseases.intersection(dsapp_diseases)
print("Common diseases:", common_diseases)
print("Number of common diseases:", len(common_diseases))

print("\nAre DSAPP’s 4 symptoms (e.g., 'Fever', 'Cough') in SymbiPredict’s 132?=================")
dsapp_symptoms = ['Fever', 'Cough', 'Fatigue', 'Difficulty Breathing']
symb_symptoms = symb.columns[:-1]  # Exclude 'prognosis'
common_symptoms = [sym for sym in dsapp_symptoms if sym.lower() in [s.lower() for s in symb_symptoms]]
print("Common symptoms:", common_symptoms)

print("\nIs SymbiPredict balanced across its 41 conditions?=====================================")
print(symb['prognosis'].value_counts())
print("Are DSAPP’s outcomes ('Positive'/'Negative') balanced?=================================")
print(dsapp['Outcome Variable'].value_counts())
print("\nrange of 'Age' or distribution of 'Gender' in DSAPP===================================")
print(dsapp['Age'].describe())
print(dsapp['Gender'].value_counts())
```

```python
import xml.etree.ElementTree as ET
tree = ET.parse('mplus_topics_2025-06-24.xml')
root = tree.getroot()
conditions = []
for topic in root.findall('topic'):
    name = topic.find('title').text
    definition = topic.find('definition').text if topic.find('definition') is not None else ''
    conditions.append({'Disease': name, 'Definition': definition})  # Add more fields
medline_df = pd.DataFrame(conditions)
```

```python
# trying to merge symb and dsapp
common_diseases = {'Chickenpox', 'Hepatitis B', 'Hypoglycemia', 'Hypothyroidism', 'Migraine', 'Common Cold', 'Osteoarthritis', 'Hyperthyroidism', 'Malaria', 'Psoriasis', 'Urinary Tract Infection', 'Tuberculosis', 'Gastroenteritis', 'Acne', 'Pneumonia'}
symb_filtered = symb[symb['prognosis'].isin(common_diseases)]
dsapp_filtered = dsapp[dsapp['Disease'].isin(common_diseases)]
dsapp_filtered.loc[:, 'Cough'] = dsapp_filtered['Cough'].map({'Yes': 1, 'No': 0})
dsapp_filtered.loc[:, 'Fatigue'] = dsapp_filtered['Fatigue'].map({'Yes': 1, 'No': 0})
merged = pd.concat([symb_filtered, dsapp_filtered[['Disease', 'Cough', 'Fatigue', 'Age', 'Gender', 'Outcome Variable']].rename(columns={'Disease': 'prognosis'})], axis=0)
merged.to_csv('merged_symptom_checker.csv', index=False)
```
